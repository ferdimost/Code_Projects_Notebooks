#Risetime/falltime

```python
A = np.zeros_like(np.loadtxt('399hz1_01.CSV',skiprows=2,delimiter=',').T)
for i in range(1,32):
    if i<10:
        si='0'+str(i)
    else:
        si=str(i)
    A+=np.loadtxt('399hz1_'+si+'.CSV',skiprows=2,delimiter=',').T
A/=i
```

```python
def ex(x,a,b,c):
    return a*np.exp((x-c)/b)
def ex2(x,a,b,c,d):
    return d+a*np.exp((x-c)/b)
```

```python
a=715
e=725
X=np.linspace(A[0,a],A[0,e-1],1000)
plt.plot(A[0,a:e],A[1,a:e])
fit=spo.curve_fit(ex,A[0,a:e],A[1,a:e],[300,1,51])[0]
plt.plot(X,ex(X,fit[0],fit[1],fit[2]))
print(fit[1])
```

```python
a=725
e=-1
X=np.linspace(A[0,a],A[0,e-1],1000)
plt.plot(A[0,a:e],A[1,a:e])
fit=spo.curve_fit(ex,A[0,a:e],A[1,a:e],[300,-1000,52])[0]
plt.plot(X,ex(X,fit[0],fit[1],fit[2]))
print(fit[1])
```

```python
a=725
e=950
X=np.linspace(A[0,a],A[0,e-1],1000)
plt.plot(A[0,a:e],A[1,a:e])
fit=spo.curve_fit(ex2,A[0,a:e],A[1,a:e],[300,-100,52,270])[0]
plt.plot(X,ex2(X,fit[0],fit[1],fit[2],fit[3]))
print(fit[1])
```
