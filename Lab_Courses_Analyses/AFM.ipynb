```python
X=np.linspace(gold1.T[0,0],gold1.T[0,255],256)
plt.plot(gold1.T[0,0:255],Average_back_forward_gold_1[0:255],color='b',label='measurement data gold1')
plt.plot(gold2.T[0,0:255],Average_back_forward_gold_2[0:255],color='b',label='measurement data gold2')
fit1=spo.curve_fit(parabola,gold1.T[0,0:255],Average_back_forward_gold_1[0:255],[8.5*10**(-3),8*10**(-5),8*10**(-5)])[0]
fit2=spo.curve_fit(parabola,gold2.T[0,0:255],Average_back_forward_gold_2[0:255],[8.5*10**(-3),8*10**(-5),8*10**(-5)])[0]
fit3=spo.curve_fit(parabola,gold3.T[0,0:255],Average_back_forward_gold_3[0:255],[8.5*10**(-3),8*10**(-5),8*10**(-5)])[0]
fit4=spo.curve_fit(parabola,gold4.T[0,0:255],Average_back_forward_gold_4[0:255],[8.5*10**(-3),8*10**(-5),8*10**(-5)])[0]
fit5=spo.curve_fit(parabola,gold5.T[0,0:255],Average_back_forward_gold_5[0:255],[8.5*10**(-3),8*10**(-5),8*10**(-5)])[0]
fit6=spo.curve_fit(parabola,gold6.T[0,0:255],Average_back_forward_gold_6[0:255],[8.5*10**(-3),8*10**(-5),8*10**(-5)])[0]
fit7=spo.curve_fit(parabola,gold7.T[0,0:255],Average_back_forward_gold_7[0:255],[8.5*10**(-3),8*10**(-5),8*10**(-5)])[0]
fit8=spo.curve_fit(parabola,gold8.T[0,0:255],Average_back_forward_gold_8[0:255],[8.5*10**(-3),8*10**(-5),8*10**(-5)])[0]
fit9=spo.curve_fit(parabola,gold9.T[0,0:255],Average_back_forward_gold_9[0:255],[8.5*10**(-3),8*10**(-5),8*10**(-5)])[0]
plt.plot(X,parabola(X,fit1[0],fit1[1],fit1[2]),color='red',label="fit curve gold1")
plt.plot(X,parabola(X,fit2[0],fit2[1],fit2[2]),color='red',label="fit curve gold2")
plt.xlabel('Bias Voltage [V]',fontsize=20)
plt.ylabel('Amplitude [V]',fontsize=20)
plt.legend(loc='upper right',fontsize=12)
plt.savefig("gold_exempl.pdf",bbox_inches="tight",pad_inches=0.05)
plt.show()
print(fit1[0],fit1[1],fit1[2])
print(fit2[0],fit2[1],fit2[2])
def parabola1(x):
    return fit1[0]*x**2+fit1[1]*x+fit1[2]
def parabola2(x):
    return fit2[0]*x**2+fit2[1]*x+fit2[2]
def parabola3(x):
    return fit3[0]*x**2+fit3[1]*x+fit3[2]
def parabola4(x):
    return fit4[0]*x**2+fit4[1]*x+fit4[2]
def parabola5(x):
    return fit5[0]*x**2+fit5[1]*x+fit5[2]
def parabola6(x):
    return fit6[0]*x**2+fit6[1]*x+fit6[2]
def parabola7(x):
    return fit7[0]*x**2+fit7[1]*x+fit7[2]
def parabola8(x):
    return fit8[0]*x**2+fit8[1]*x+fit8[2]
def parabola9(x):
    return fit9[0]*x**2+fit9[1]*x+fit9[2]
minimum1=fmin(parabola1,0.3)[0]
minimum2=fmin(parabola2,0.3)[0]
minimum3=fmin(parabola3,0.3)[0]
minimum4=fmin(parabola4,0.3)[0]
minimum5=fmin(parabola5,0.3)[0]
minimum6=fmin(parabola6,0.3)[0]
minimum7=fmin(parabola7,0.3)[0]
minimum8=fmin(parabola8,0.3)[0]
minimum9=fmin(parabola9,0.3)[0]
Minima=[minimum1,minimum2,minimum3,minimum4,minimum5,minimum6,minimum7,minimum8,minimum9]
print(Minima)
print(np.mean(Minima),np.std(Minima))
```
